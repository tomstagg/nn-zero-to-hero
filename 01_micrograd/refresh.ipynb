# Closures

A closure is a function that captures variables from the scope where it was defined,
even after that scope has finished executing.

```python
def make_multiplier(factor):
    def multiply(x):
        return x * factor  # captures `factor` from the outer scope
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))   # 10
print(triple(5))   # 15
# `factor` is gone from make_multiplier's scope, but multiply still holds it
```

# Value object

A minimal Value class. Each Value wraps a number and tracks:
- `data` — the scalar
- `_backward` — a closure that knows how to propagate gradients back through this operation

```python
class Value:
    def __init__(self, data):
        self.data = data
        self.grad = 0.0
        self._backward = lambda: None  # default: do nothing

    def __mul__(self, other):
        out = Value(self.data * other.data)

        # closure: captures self, other, out
        def _backward():
            self.grad += other.data * out.grad   # d(self) = other * grad
            other.grad += self.data * out.grad   # d(other) = self * grad
        out._backward = _backward

        return out

    def __repr__(self):
        return f"Value(data={self.data}, grad={self.grad})"


a = Value(3.0)
b = Value(4.0)
c = a * b          # c.data = 12.0, _backward captures a, b, c

c.grad = 1.0       # seed gradient at the output
c._backward()      # propagate back

print(a)           # grad should be 4.0 (dc/da = b)
print(b)           # grad should be 3.0 (dc/db = a)
```
